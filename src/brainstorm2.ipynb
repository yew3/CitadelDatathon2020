```python
import numpy as np
import pandas as pd
import os
import sys
from pathlib import Path
from datetime import datetime, timezone, timedelta
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# CONSTANTS
path = os.getcwd()   
DATA_DIR = os.path.abspath(os.path.join(path, os.pardir, "data"))
MOVIE_LENS_DIR = os.path.abspath(os.path.join(path,  os.pardir, "data", "movie_lense"))
NETFLIX_DIR = os.path.abspath(os.path.join(path,  os.pardir, "data", "netflix"))
MOVIE_DIR = os.path.abspath(os.path.join(path,  os.pardir, "data", "ml-1m"))
print(DATA_DIR)
```

```python
users_df = pd.read_csv(os.path.join(MOVIE_DIR, "users.dat"), header=None, names=["user_id", "sex", "age", "occupation", "zipcode"], delimiter= "::")
users_movies_df = pd.read_csv(os.path.join(MOVIE_DIR, "movies.dat"), header=None, names=["movie_id", "title", "genre"], delimiter= "::")
users_ratings_df = pd.read_csv(os.path.join(MOVIE_DIR, "ratings.dat"), names=["user_id", "movie_id", "rating", "timestamp"], header=None, delimiter= "::")

industry_df = pd.read_csv(os.path.join(DATA_DIR, "movie_industry.csv"), encoding='latin-1')
oscar_df = pd.read_csv(os.path.join(DATA_DIR, "the_oscar_award.csv"))
oscar_df = pd.read_csv(os.path.join(DATA_DIR, "Oscars-demographics-DFE.csv"), encoding = 'latin-1')
uszips_df = pd.read_csv(os.path.join(DATA_DIR, "uszips.csv"), dtype=str)
uszips_df.rename(columns={"zip": "zipcode"})

score_df = pd.read_csv(os.path.join(MOVIE_LENS_DIR, "genome-scores.csv"))
genome_tag_df = pd.read_csv(os.path.join(MOVIE_LENS_DIR, "genome-tags.csv"))
ratings_df = pd.read_csv(os.path.join(MOVIE_LENS_DIR, "ratings.csv"))
movies_df = pd.read_csv(os.path.join(MOVIE_LENS_DIR, "movies.csv"))
tags_df = pd.read_csv(os.path.join(MOVIE_LENS_DIR, "tags.csv"))
links_df = pd.read_csv(os.path.join(MOVIE_LENS_DIR, "links.csv"))
```

```python
uszips_df = uszips_df.rename(columns={"zip": "zipcode"})
```

```python
merge_users_df = pd.merge(users_ratings_df, users_df, how='inner', on="user_id")
```

```python
zc_movie_rating_df = merge_users_df.groupby(
   ['movie_id', 'zipcode']
).agg(
    {
        'rating':'std',    # standard deviation of ratings
    }
).dropna()
```

```python
zc_rating_df = zc_movie_rating_df.groupby(
    ['zipcode']
).agg(
    {
        'rating':'mean',
    }
)
```

```python
pd.merge(zc_rating_df, uszips_df[['zipcode','city','state_id']], how="inner", on="zipcode").sort_values(by=['rating'],ascending=False).head(20)
```
